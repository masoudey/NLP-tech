# NER and Sentiment Analysis Visualizations

This notebook loads evaluation metrics and prediction outputs from the NER and sentiment analysis scripts, and visualizes them using tables and charts. Make sure you have saved your script outputs as CSV files (e.g., `ner_metrics.csv`, `ner_predictions.csv`, `sentiment_metrics.csv`, `sentiment_predictions.csv`) in the working directory.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load evaluation metrics
ner_metrics = pd.read_csv('ner_metrics.csv')  # columns: precision, recall, f1, accuracy
sent_metrics = pd.read_csv('sentiment_metrics.csv')  # columns: accuracy

# Display metrics tables
display(ner_metrics)
display(sent_metrics)

# Plot NER metrics
fig, ax = plt.subplots()
ner_metrics.plot.bar(ax=ax)
ax.set_title('NER Evaluation Metrics')
ax.set_xlabel('Metric')
ax.set_ylabel('Score')
plt.tight_layout()
plt.show()

# Plot Sentiment Analysis Accuracy
fig, ax = plt.subplots()
ax.bar(['Sentiment Accuracy'], sent_metrics['accuracy'])
ax.set_ylim(0, 1)
ax.set_title('Sentiment Analysis Accuracy')
plt.tight_layout()
plt.show()
```
